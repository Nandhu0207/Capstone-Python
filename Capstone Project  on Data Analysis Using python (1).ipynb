# task 7

```python
correlation = movie['popularity'].corr(movie['budget'])
```

```python
if correlation > 0:
    interpretation = "There is a positive correlation between movie popularity and budget."
elif correlation < 0:
    interpretation = "There is a negative correlation between movie popularity and budget."
else:
    interpretation = "There is no linear correlation between movie popularity and budget."

print("Correlation Coefficient:", correlation)
print("Interpretation:", interpretation)
```

Interpretation of the correlation coefficient:

If the correlation is close to 1, it suggests a strong positive linear relationship, meaning that as movie budgets increase, popularity tends to increase as well.

If the correlation is close to -1, it suggests a strong negative linear relationship, meaning that as movie budgets increase, popularity tends to decrease.

If the correlation is close to 0, it suggests no significant linear relationship between the two variables.

The interpretation will give you insights into whether there is a correlation between movie popularity and budget and the direction of that correlation. Keep in mind that correlation does not imply causation, and other factors may also influence movie popularity.

```python
movie.columns
```

```python
top_500_movies = movie.sort_values(by='revenue', ascending=False).head(500)

# Calculate measures of central tendency for 'Budget', 'Revenue', and 'Runtime'
mean_budget = top_500_movies['budget'].mean()
median_budget = top_500_movies['budget'].median()
mean_revenue = top_500_movies['revenue'].mean()
median_revenue = top_500_movies['revenue'].median()
mean_runtime = top_500_movies['runtime'].mean()
median_runtime = top_500_movies['runtime'].median()
```

```python
print("Measures of Central Tendency for Top 500 Movies:")
print(f"Budget: Mean = {mean_budget}, Median = {median_budget}")
print(f"Revenue: Mean = {mean_revenue}, Median = {median_revenue}")
print(f"Runtime: Mean = {mean_runtime}, Median = {median_runtime}")
```

#### Box Plot for Budget

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.boxplot(top_500_movies['budget'])
plt.title('Budget Box Plot')
```

#### Box Plot for Revenue

```python
plt.subplot(1, 3, 2)
plt.boxplot(top_500_movies['revenue'])
plt.title('Revenue Box Plot')
```

#### Box Plot for Runtime

```python
plt.subplot(1, 3, 3)
plt.boxplot(top_500_movies['runtime'])
plt.title('Runtime Box Plot')

plt.tight_layout()
plt.show()
```
